# Data Cleaning

```python
import glob
import os
import sys

import pandas as pd

sys.path.append(os.path.abspath(".."))
from src.data_cleaning import (clean_date, extract_language,
                               keep_relevant_columns)
```

```python
file_paths = glob.glob('../data/1_bronze/*.csv')

hospital_dfs = {}

for file_path in file_paths:
    hospital_name = os.path.basename(file_path).replace('.csv', '')
    df = pd.read_csv(file_path)
    print(f"\n-------------- {hospital_name} -------------- \n")
    display(df.head())
    
    hospital_dfs[hospital_name] = df
```

We will clean data by:
- keeping only relevant columns: usernames, review date and review content
- adding hospital name so that we can merge reviews of the two hospitals together
- keeping only reviews with content (so, excluding those without text), as we are doing text-based analysis
- replacing 'Edited ' string in the date so that we can parse the date
- parsing the date and keeping only year, month and date
- merging reviews for the two hospitals together

```python
name_map = {
    "uz_brussel": "UZ Brussel",
    "uz_leuven_pellenberg": "UZ Leuven Pellenberg",
    "uz_leuven_gasthuisberg": "UZ Leuven Gasthuisberg",
    "uz_gent": "UZ Gent",
    "uz_antwerpen": "UZ Antwerpen"
}
 
cleaned_dfs = []

for hospital_key, df in hospital_dfs.items():
    hospital_name = name_map.get(hospital_key, hospital_key)

    df_clean = keep_relevant_columns(df)

    df_clean["hospital"] = hospital_name

    df_clean["original_language"] = df_clean["original_language"].apply(extract_language)

    df_clean = df_clean[df_clean["content"].notna()]

    df_clean["year"] = clean_date(df_clean["year"], reference_date="2025-07-10")

    cleaned_dfs.append(df_clean)

combined_df = pd.concat(cleaned_dfs, ignore_index=True)
```
